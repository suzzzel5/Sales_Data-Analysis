```python
df["TotalAmount"] = df["Quantity"] * df["Price"]  #Totalamount vanni euta coloumn banayo "varaible" tesma each quantity ko price seta multiply gareko to know the TA ko each 
```

```python
df[df["TotalAmount"] > 30000] #Order above 30k . This returns ROWS, not True/False.
```

**groupby() means "Split->Calculate->Combine"**


```python
#Grouping (CORE ANALYST SKILl)
df.groupby("City")["TotalAmount"].sum() #City ma kati ko Total sales hunxa vanera we use groupby / () call or do something
                                        #() for the work to do " run action " / [] to pick the data from coloumn "select data"
```

```python
df.groupby("Category")["TotalAmount"].sum() #kun Category ko product total sales bhayo vanera 
```

```python
df.groupby("City")["TotalAmount"].sum().sort_values(ascending=False) #this shows which city sells the most item by totalamount sales
```

```python
df.groupby("Product")["TotalAmount"].sum().sort_values(ascending=True) #sort_value arrange the reslults in order based on values from smallest to largest (ascending=True)small to big (ascending=false)big to small
#This tells you which product earns the most money.
```
