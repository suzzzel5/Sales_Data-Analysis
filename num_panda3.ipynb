```python
import pandas as pd
```

```python
df = pd.read_csv("sales_backup.csv")
print(df.to_string())
```

```python
#Customer Behavior analysis   /// for each customername add orderID
#Count Orders per customer
df.groupby("CustomerName")["OrderID"].count() #Ekjana customer le kati ota oder place garyo vanera
```

```python
#How much did one person spend total
df.groupby("CustomerName")["NetAmount"].sum().sort_values(ascending=False) 
```

 So, we can see that Anish is the Top buyer who have purchased the most . and suman is the least 


```python
#Quantity bought by each customer
df.groupby("CustomerName")["Quantity"].sum().sort_values(ascending=False) #Ekjana le kati ota quantity purachse garyo ("Kosly dherai saman buy garyo")
```

# Pattern to Remember
Customer analysis always follows this form.

Count orders.

Sum spending.

Sum quantity.

```python
#Showing the top customer who have purchased more than 50k 
top_customer = df.groupby("CustomerName")["NetAmount"].sum().sort_values(ascending=False)
top_customer[top_customer > 50000]  # yedi top_customer < 50000 halyo vane it give T/F so, aghadi top_customer to display the amount
```

# Charts

```python
top_customer.sort_values(ascending=False).plot(kind="bar") #showing top customer on top to buttom
```

```python
#City revenue chart
df.groupby("City")["NetAmount"].sum().plot(kind="bar") # Showing which city purchsed the most revenue
```

```python
df.groupby("Category")["NetAmount"].sum().sort_values(ascending=False).plot(kind="bar")
```
