
# Charts Visualization 


```python
import pandas as pd
```

```python
df = pd.read_csv("sales_backup_2.csv")
print(df.to_string())
```

# Using matplotlib.pylot for charts

```python
import matplotlib.pyplot as plt
```

```python
category_sales = df.groupby("Category")["NetAmount"].sum().sort_values(ascending=False)
category_sales.plot(kind="bar")

plt.title("Net Revenue by Category")  #Title / heading 
plt.xlabel("Categroy")                 #x-axis showing each category
plt.ylabel("Net Amount")               #y-axis showing Net Amounts
plt.show()
```

```python
city_sales = df.groupby("City")["NetAmount"].sum().sort_values(ascending=False)
city_sales.plot(kind="bar")

plt.title("City sales Net Amount")
plt.xlabel("City")
plt.ylabel("Net Amount")
plt.show()
```

Kathmandu city contributes the most revenue where as latipur contributes the least 

```python
product_sales = df.groupby("Product")["NetAmount"].sum().sort_values(ascending=False)
product_sales.plot(kind="bar")

plt.title("Product sales Net Amount")
plt.xlabel("Proudcts")
plt.ylabel("Net Amount")

plt.show()
```

The sales of laptop is higher 

```python
#Quantity vs Revenue thinking
product_quantity = df.groupby("Product")["Quantity"].sum()
product_quantity.plot(kind="bar")

plt.title("Total Quantity sold by products")
plt.xlabel("Products")
plt.ylabel("Quantity")

plt.show()

```
